procesing data

```python
class Pemain:
    def __init__(self, nama, usia, nilai, harga, posisi):
        self.name = nama
        self.age = usia
        self.overall = nilai
        self.value = harga
        self.position = posisi
```

```python
def cari_pemain(pemain_fifa, gent, terbaik=False):
    gent_data = pemain_fifa.sort_values(['Overall'], ascending=False)
    if terbaik:
        pemain = 0
    else:
        pemain = random.randint(0, len(gent_data)-1)
    item_pemain = Pemain(
        gent_data.iloc[pemain]['Name'], 
        gent_data.iloc[pemain]['Age'],
        gent_data.iloc[pemain]['Overall'],
        gent_data.iloc[pemain]['Value'],
        gent_data.iloc[pemain]['Position']
    )
    
    return item_pemain
```

```python
def masing_masing_pemain(bola_data):
    gen_data = list()
    posisi = [
        'Goalkeeper', 'Left Defender', 'Right Defender', 
        'Center Defender', 'Left Half', 'Right Half', 
        'Center Half', 'Left Forward', 'Right Forward', 'Center Forward'
    ] 
    
    for posisi in posisi:
        pemain_df = bola_data[bola_data['Position'] == posisi]
        for i in range(2):
            item_pemain = cari_pemain(pemain_df, gen_data)
            gen_data.append(item_pemain)
        if posisi == 'Center Defender':
            for i in range(2):
                item_pemain = cari_pemain(pemain_df, gen_data)
                gen_data.append(item_pemain)

    return gen_data
```

```python
def fungsi_fitness(maks_uang, rata_rata_usia, individu):
    nilai = 0
    harga = 0
    usia = 0
    for pemain in individu:
        nilai += pemain.overall
        harga += pemain.value
        usia += pemain.age

    nilai += min(0, maks_uang - harga)
    usia_hasil = (rata_rata_usia - usia / 22) * 10
    nilai += min(0, usia_hasil)
    
    if len(list(set([it.name for it in individu]))) != 22:
        nilai -= 1000000
        
    return nilai
```

```python
def mutasi(individu, peluang_mutasi, peluang_terbaik_mutasi, data_bola):
    data_baru = list()
    for generate in individu:
        pemain_fifa = data_bola[data_bola['Position'] == generate.position]
        if peluang_terbaik_mutasi > random.random():
            item_pemain = cari_pemain(pemain_fifa, individu, terbaik=True)
            data_baru.append(item_pemain)
            continue
        if peluang_mutasi > random.random():
            item_pemain = cari_pemain(pemain_fifa, individu, terbaik=True)
            data_baru.append(item_pemain)
        else:
           item_pemain = cari_pemain(pemain_fifa, individu, terbaik=True)
           data_baru.append(item_pemain)
    return data_baru
```

```python
maksimal_uang = 1500000000
usia_rata_rata = 25
Epochs = 100
anak = 25
Peluang_Mutasi = 0.001
Peluang_Putasi_Terbaik = 0.01
Epochs_Change_Over_Mutasi = 100
Mutasi_Berkurang = 0.000001
INCREASE_IF_NO_IMPROVES = 90
Mutasi_Bertambah = 0.0001
Anak_Bertambah = 3
```

```python
ind = masing_masing_pemain(data_fifa)
stable_score = 0
best_score = fungsi_fitness(maksimal_uang, usia_rata_rata, ind)
best_ind = ind
for i in range(Epochs):
    improve_flag = False
    if stable_score == INCREASE_IF_NO_IMPROVES:
        stable_score = 0
        print('Peluang mutasi bertambah menjadi: ', Peluang_Mutasi + Mutasi_Bertambah)
        Peluang_Mutasi += Mutasi_Bertambah
        anak += Anak_Bertambah
    if i % 5 == 0:
        print('Indeks: ', i,'Score Terbaik: ', best_score)
    if i % Epochs_Change_Over_Mutasi == 0 and i != 0:
        print('Peluang mutasi berkurang menjadi: ', Peluang_Mutasi - Mutasi_Berkurang)
        Peluang_Mutasi -= Mutasi_Berkurang
    for i in range(anak):
        child = mutasi(ind, Peluang_Mutasi, Peluang_Putasi_Terbaik, data_fifa)
        child_score = fungsi_fitness(maksimal_uang, usia_rata_rata, child)
        if child_score >= best_score:
            best_score = child_score
            best_ind = child
            improve_flag=True
            
    if improve_flag == False:
        stable_score += 1
```
